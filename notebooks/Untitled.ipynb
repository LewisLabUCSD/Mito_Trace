```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def half_ag(x):
    return x/2 + 7


x = np.arange(10, 100)
y = [half_ag(i) for i in x]
f, ax = plt.subplots(figsize=(16,16))
plt.scatter(x, y)
plt.xlabel("age")
plt.ylabel("datee")
plt.xticks(np.arange(min(x), max(x)+1, 1.0), rotation=90);
plt.yticks(np.arange(min(y), max(y)+1, 1.0));
```
