```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu = 998.8 
sigma = 73.10
x1 = 900
x2 = 1100

# calculate the z-transform
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma

x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
# mean = 0, stddev = 1, since Z-transform was calculated
y2 = norm.pdf(x_all,0,1)
```

```python
mu1 = 1
mu2 = 0
sigma = 0.5

x = np.arange(-3,3,0.001)
y1 = norm.pdf(x,mu1,sigma)
y2 = norm.pdf(x,mu2,sigma)
```

```python
import matplotlib as mpl
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.style.use('fivethirtyeight')
mpl.rcParams.update({'font.size': 22})

ax.plot(x,y1)
ax.plot(x,y2,color='g')
plt.legend(["Donors with CHIP", "Donors without"])
ax.set_xlabel('Change in cluster size distribution')
ax.set_yticklabels([])
ax.set_title('Cluster imbalance entropy')

plt.savefig('CHIP_status_cluster_entropy.svg',transparent=True, bbox_inches='tight')
```
