```python
import os
from os.path import basename, exists, join
import pandas as pd
from src.config import ROOT_DIR

os.chdir(ROOT_DIR)
```

```python
outdir = "data/processed/2020_09_07_Croker_mito"
outdir_cov = join(outdir, "coverage")
if not exists(outdir):
    os.mkdir(outdir)
    
if not exists(outdir_cov):
    os.mkdir(outdir_cov)
```

```python
samples = pd.read_csv("parameters/2020_09_07_Croker_mito.csv")
samples
```

```python
for ind, val in samples.iterrows():
    #cmd = f"samtools depth  -r MT -Q 20  > {outdir_cov}/{val['sample']}_coverage_Q20_MT.tsv"
    
    cmd = f"bamCoverage -b {val['bam_f']} -of bedgraph -p 10 -o {outdir_cov}/{val['sample']}.bedgraph"
    print(cmd)
    !{cmd}
```

## Normalize to CPM

```python
for ind, val in samples.iterrows():
    #cmd = f"samtools depth  -r MT -Q 20  > {outdir_cov}/{val['sample']}_coverage_Q20_MT.tsv"
    
    cmd = f"bamCoverage -b {val['bam_f']} --normalizeUsing CPM -of bedgraph -p 10 -o {outdir_cov}/CPM_{val['sample']}.bedgraph"
    print(cmd)
    !{cmd}
```

## Different strands

```python
for ind, val in samples.iterrows():
    #cmd = f"samtools depth  -r MT -Q 20  > {outdir_cov}/{val['sample']}_coverage_Q20_MT.tsv"
    
    cmd = f"bamCoverage --filterRNAstrand forward -b {val['bam_f']} -of bedgraph -p 10 -o {outdir_cov}/{val['sample']}_for.bedgraph"
    print(cmd)
    !{cmd}
    
for ind, val in samples.iterrows():
    #cmd = f"samtools depth  -r MT -Q 20  > {outdir_cov}/{val['sample']}_coverage_Q20_MT.tsv"
    
    cmd = f"bamCoverage --filterRNAstrand reverse -b {val['bam_f']} -of bedgraph -p 10 -o {outdir_cov}/{val['sample']}_rev.bedgraph"
    print(cmd)
    !{cmd}
```
