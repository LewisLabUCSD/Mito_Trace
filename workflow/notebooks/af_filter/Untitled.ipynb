```python
import pandas as pd
from os.path import join, dirname
```

```python
# mult_cells = snakemake.input.mult_cells
# cov_dirs = snakemake.input.cov_dirs
# samples = snakemake.params.samples
# out_merge_pile = snakemake.output.merge_pile


cov_dirs = ["/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/Flt3l/MT/cellr_True/numread_200/",
           "/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/Input/MT/cellr_True/numread_200/"]
mult_cells = ["/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/merged/MT/cellr_True/numread_200/filters/minC10_minR50_topN0_hetT0.001_hetC10_hetCount5_bq20/mgatk/vireoIn/multiplex/cells_meta.tsv"]
samples = ["Flt3l", "Input"]
out_merge_pile = ["/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/coverage_merged/donor0/MT/cellr_True/numread_200/merge.coverage.strands.txt",
              "/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/coverage_merged/donor1/MT/cellr_True/numread_200/merge.coverage.strands.txt",
             ]
```

```python
cells_meta = pd.read_csv(mult_cells[0], sep="\t")
from collections import defaultdict
donor_cov_d = defaultdict(list)
for d, cells_don_df in cells_meta.groupby("donor"):
    for n in ["A", "G", "C", "T", "coverage"]:
        print('n', n)
        curr_pile = []
        for ind, curr_sample in enumerate(samples):
            print(curr_sample)
            cov_df = pd.read_csv(join(dirname(cov_dirs[ind]), f"{curr_sample}.{n}.strands.txt"), header=None)
            cov_df[1] = cov_df[1].apply(lambda x: f"{x}_{curr_sample}")
            # filter for cells in this donor
            curr_don_cov = cov_df.loc[cov_df[1].isin(cells_don_df["ID"])].copy()
            print(cov_df.head())
            print(curr_don_cov.head())
            donor_cov_d[(d,n)].append(curr_don_cov)
        #curr_pile.append(cov_df)

```

```python
for k in donor_cov_d:
    d,n = k[0], k[1]
    print('d,n', d, n)
    curr_df = pd.concat(donor_cov_d[k], axis=0, ignore_index=True)
    print(curr_df.head())
    (curr_df).to_csv(join(dirname(out_merge_pile[int(d)]), f"merge.{n}.strands.txt"), sep="\t", index=False)
```
