```r
all_f <- list()
top_umap_clust_col_str = "" #list of files separated by space to be concatenated later
for (mt in (top.clones)){
    print(mt)    
    #p <- FeaturePlot(se,features=mt, split.by = "condition", order=T) & theme(legend.position = c(0.1,0.2))
    #curr.count = sum(se[[mt]]>0, na.rm=TRUE)
    #ggsave(file.path(outdir, paste("withNA.count", curr.count,mt, "groupCond.png", sep="_")), p)
    
    # Run with removing NA 
    curr.id <- paste0("cloneID_",mt)
    expr <- FetchData(se, vars = curr.id)
    print(head(expr))
    #if(length(which((!is.na(expr)) & (expr=0)) ) != 0){
    try({
        curr.se <- se[, which((!is.na(expr)) & (expr!=0))]
    #     p <- FeaturePlot(curr.se, features=curr.id, split.by=c("clusterID"), shape.by=c("condition"), pt.size=(top_clones/8),
    #                      order=T) & theme(legend.position = c(0.1,0.2))
    #     curr.count = sum(curr.se[[curr.id]]>0, na.rm=TRUE)

        p <- DimPlot(object = curr.se, label = FALSE, combine=T,
                     group.by=c("clusterID"), shape.by=c("condition"), pt.size=(top_clones/8)) +
        xlim(-1*coord-0.01,coord+0.01) +
        ylim(-1*coord-0.01,coord+0.01) + 
        theme(text = element_text(size = 4*top_clones)) +
        theme(axis.text = element_text(size = (top_clones)))
        curr.f = file.path(outdir, paste0("top_umap_clust_col_",mt,".svg"))
        
        ggsave(curr.f, device='svg',
           plot=p,width=2*top_clones, height=8*top_clones, units="in", limitsize = FALSE)

        top_umap_clust_col_str = paste(top_umap_clust_col_str, curr.f, sep=" ")
        #ggsave(file.path(outdir, "top_umap_clust_col.pdf"), plot=all_f,width=2*top_clones, height=8*top_clones, units="in", limitsize = FALSE)

        #ggsave(file.path(outdir, paste("count", curr.count,mt, "groupCond.png", sep="_")), p)
        all_f <- c(all_f,p)
        
    })
}
    
```
