## Average over clusters

```r
avg_expr <- AverageExpression(
  se,
  assays = 'RNA',
  return.seurat = FALSE,
  group.by = "cluster_labels",
  add.ident = NULL,
  slot = "data",
  verbose = TRUE,
)
```

```r
avg_expr_count <- AverageExpression(
  se,
  assays = 'RNA',
  return.seurat = FALSE,
  group.by = "cluster_labels",
  add.ident = NULL,
  slot = "counts",
  verbose = TRUE,
)
```

```r
head(avg_expr_count$RNA)
```

```r
write.table(avg_expr$RNA, file = out_data_f, 
            append = FALSE, quote = FALSE, sep = "\t",
            eol = "\n", na = "NA", dec = ".", row.names = TRUE,
            col.names = TRUE, qmethod = c("escape", "double"),
            fileEncoding = "")
```

```r
write.table(avg_expr_count$RNA, file = out_counts_f, 
            append = FALSE, quote = FALSE, sep = "\t",
            eol = "\n", na = "NA", dec = ".", row.names = TRUE,
            col.names = TRUE, qmethod = c("escape", "double"),
            fileEncoding = "")
```
