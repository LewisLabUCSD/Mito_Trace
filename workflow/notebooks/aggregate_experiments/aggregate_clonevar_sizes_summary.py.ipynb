```python
import pandas as pd
from os.path import join, dirname, basename
import os
import numpy as np
from icecream import ic

#import seaborn as sns
```

```python
cells_meta_f = "/data/Mito_Trace/output/aggregate/CHIP_aggr/v04/data/clone_var/variants.init_knnRes.30_objs_standard/aggregate/cells_meta.tsv"
out_cl_sizes_f = "/data/Mito_Trace/output/aggregate/CHIP_aggr/v04/data/clone_var/variants.init_knnRes.30_objs_standard/aggregate/clone_sizes.tsv"
out_cl_order_f = "/data/Mito_Trace/output/aggregate/CHIP_aggr/v04/data/clone_var/variants.init_knnRes.30_objs_standard/aggregate/cloneOrder.tsv"
top_n = 6
```

```python
cells_meta = pd.read_csv(cells_meta_f, sep="\t", index_col=0)
cells_meta.head()
```

```python
clone_counts = cells_meta.groupby(["name", "donor"]).size().to_frame('ncells').reset_index() #.apply(lambda x: )
assert(not clone_counts.duplicated(subset="name").any())
clone_counts = clone_counts.set_index("name")
clone_counts.head()
```

```python
clone_ranks = clone_counts.groupby(["donor"]).apply(lambda x: x[["ncells"]].rank(method='first', ascending=False)).rename({"ncells": "rank_ncells"}, axis=1)
clone_ranks
```

```python
clone_count_rank = pd.concat((clone_counts, clone_ranks), axis=1, ignore_index=False)
clone_count_rank
```

## Save ordered version by rank, and filtered for top_n in each donor

```python
clone_count_rank = clone_count_rank.sort_values("rank_ncells")
clone_count_rank.to_csv(out_cl_sizes_f, sep="\t")
clone_count_rank.head()
```

```python
with open(out_cl_order_f, 'w') as f:
    cl_filt = clone_count_rank[clone_count_rank["rank_ncells"] <= top_n].index.values
    f.write("\n".join(cl_filt))
```
