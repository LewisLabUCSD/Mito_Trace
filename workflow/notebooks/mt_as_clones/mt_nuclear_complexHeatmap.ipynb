```r
library(ComplexHeatmap)
library(dplyr)
```

```r
af <- read.csv(af_f, row.names=1)
af
```

```r
labels <- read.csv(labels_f, row.names=1)
labels
```

```r
#sum(is.na(af))
af <- Filter(function(x)!all((is.na(x) | x==0)), af)
af <- af[rowSums((is.na(af) | af==0)) != ncol(af), ]


labels <- labels[row.names(labels) %in% row.names(af),]
```

```r
af[is.na(af)] <- 0
```

```r
af[af>af_max] <- af_max
```

```r
heat <- ComplexHeatmap::Heatmap(as.matrix(af),
        name = name,na_col = "black", row_title_rot = 0,
        row_split = labels[c("labels")],
        show_row_names = FALSE, use_raster=FALSE
       )

# heat <- tryCatch({ComplexHeatmap::Heatmap(as.matrix(af),
#         name = name,na_col = "black",
#         row_split = labels[c("labels")],
#         show_row_names = FALSE, use_raster=FALSE
#        )},
#        error = function(cond){
#         print(cond)
#         print("setting na to 0")
#         af[is.na(af)] <- 0
#         return(ComplexHeatmap::Heatmap(as.matrix(af),
#         name = name,na_col = "black",
#         row_split = labels[c("labels")],
#         show_row_names = FALSE, use_raster=FALSE)
#        )}
#        )
heat
```

```r
# Run without the big variant 
top.var <- names(which.max(colSums(af, na.rm=TRUE)))
row_ha = rowAnnotation(condition = labels$condition)

heat <- ComplexHeatmap::Heatmap(select(af, -c(top.var)), row_title_rot = 0,
        name = name,
        row_split = labels[c("labels")],
        show_row_names = FALSE, use_raster=FALSE, right_annotation=row_ha
       )
heat
```

```r
# Run without the big variant 
row_ha = rowAnnotation(condition = labels$condition)

heat <- ComplexHeatmap::Heatmap(select(af, -c(top.var)),row_title_rot = 0,
        name = name,
        show_row_names = FALSE, use_raster=FALSE, right_annotation=row_ha
       )
heat
```

```r
# ComplexHeatmap
# split by a vector
pdf((file.path(outdir, paste0(name, "_heatmap.top.pdf"))), width=8,height=8)
heat <- ComplexHeatmap::Heatmap(af,
        name = name,row_title_rot = 0,
        row_split = labels[c("labels")],
        show_row_names = FALSE, use_raster=TRUE
       )
ComplexHeatmap::draw(heat)
dev.off()
```

```r
# Run without the big variant 
row_ha = rowAnnotation(condition = labels$condition, labs = as.factor(labels$labels))

pdf((file.path(outdir, paste0(name, "_heatmap.top.noGroup.pdf"))), width=8,height=8)

#row_ha = rowAnnotation(labs = labels$labels)

heat <- ComplexHeatmap::Heatmap(af,
        name = name, right_annotation=row_ha,row_title_rot = 0,
        show_row_names = FALSE, use_raster=TRUE, left_annotation=row_ha
       )
ComplexHeatmap::draw(heat)
dev.off()
```
