```r
labels_f = "/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/merged/MT/cellr_True/numread_200/filters/minC10_minR50_topN0_hetT0.001_hetC10_hetCount5_bq20/mgatk/vireoIn/mt_as_clones/variants_simpleUnion/anno_mt_af_v02/input/minCov10_donor0.labels.csv"
af_f = "/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/merged/MT/cellr_True/numread_200/filters/minC10_minR50_topN0_hetT0.001_hetC10_hetCount5_bq20/mgatk/vireoIn/mt_as_clones/variants_simpleUnion/anno_mt_af_v02/input/minCov10_donor0.af.csv"
outdir = "/data/Mito_Trace/output/pipeline/v04/CHIP_b1/MTBlacklist_A2/data/merged/MT/cellr_True/numread_200/filters/minC10_minR50_topN0_hetT0.001_hetC10_hetCount5_bq20/mgatk/vireoIn/mt_as_clones/variants_simpleUnion/anno_mt_af_v02/input/method_bias/"
name = "donor0"
```

```r
library(ComplexHeatmap)
library(dplyr)
```

```r
af <- read.csv(af_f, row.names=1)
af
```

```r
labels <- read.csv(labels_f, row.names=1)
labels
```

```r
print(dim(labels))

print(dim(af))
```

```r
heat <- ComplexHeatmap::Heatmap(af,
        name = name,
        row_split = labels[c("seurat_clusters")],
        show_row_names = FALSE, use_raster=FALSE
       )
heat
```

```r
# Run without the big variant 
row_ha = rowAnnotation(condition = labels$condition)

heat <- ComplexHeatmap::Heatmap(select(af, -c("X5581G")),
        name = name,
        row_split = labels[c("seurat_clusters")],
        show_row_names = FALSE, use_raster=FALSE, right_annotation=row_ha
       )
heat
```

```r
# Run without the big variant 
row_ha = rowAnnotation(condition = labels$condition)

heat <- ComplexHeatmap::Heatmap(select(af, -c("X5581G")),
        name = name,
        show_row_names = FALSE, use_raster=FALSE, right_annotation=row_ha
       )
heat
```

```r
# ComplexHeatmap
# split by a vector
pdf((file.path(outdir, paste0(name, "_heatmap.top.pdf"))), width=8,height=8)
heat <- ComplexHeatmap::Heatmap(af,
        name = name,
        row_split = labels[c("seurat_clusters")],
        show_row_names = FALSE, use_raster=TRUE
       )
ComplexHeatmap::draw(heat)
dev.off()
```

```r
# Run without the big variant 
row_ha = rowAnnotation(condition = labels$condition)

pdf((file.path(outdir, paste0(name, "_heatmap.top.remove5581G.pdf"))), width=8,height=8)

heat <- ComplexHeatmap::Heatmap(select(af, -c("X5581G")),
        name = name,
        row_split = labels[c("seurat_clusters")],
        show_row_names = FALSE, use_raster=TRUE, right_annotation=row_ha
       )
ComplexHeatmap::draw(heat)
dev.off()
```
