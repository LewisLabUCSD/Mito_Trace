```r
all_f <- list()
top_umap_clust_col_str = "" #list of files separated by space to be concatenated later
for (mt in (colnames(cells_meta))){
    print(mt)    
    #p <- FeaturePlot(se,features=mt, split.by = "condition", order=T) & theme(legend.position = c(0.1,0.2))
    #curr.count = sum(se[[mt]]>0, na.rm=TRUE)
    #ggsave(file.path(outdir, paste("withNA.count", curr.count,mt, "groupCond.png", sep="_")), p)
    
    # Run with removing NA 
    #curr.id <- paste0("cloneID_",mt)
    curr.id <- mt
    expr <- FetchData(se, vars = curr.id)
    print(head(expr))
    try({
        #curr.se <- se[, which((!is.na(expr)) & (expr!=0))]
        curr.se <- se[, which((!is.na(expr)))]
        p <- FeaturePlot(curr.se, features=curr.id, shape.by=c("condition"),
                 order=T) & theme(legend.position = c(0.1,0.2)) +
                xlim(-1*coord-0.01,coord+0.01) +
                ylim(-1*coord-0.01,coord+0.01) 
        
        curr.f = file.path(outdir, paste0("np.group.top_umap_clust_col_",mt,".pdf"))
        ggsave(curr.f, plot=p, device='pdf',plot=p,width=4, height=4, units="in", limitsize = FALSE)


        top_umap_clust_col_str = paste(top_umap_clust_col_str, curr.f, sep=" ")
        all_f <- c(all_f,p)
        
    })
}
    
```
