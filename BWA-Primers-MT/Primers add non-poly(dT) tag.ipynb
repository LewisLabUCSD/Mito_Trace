## To each primer adding tag to each primer
"Protocol steps correspond to the Chromium Next GEM Single Cell V(D)J v1.1 Reagent Kits User Guide (CG000207)."

![non_poly_dt](images/non_poly_dt_tag.png)

```python
tag = "AAGCAGTGGTATCAACGCAGAGTAC"
```

```python
import pandas as pd
```

```python
df = pd.read_csv("final_primers.csv")
df["Sequence 5\' with tag"] = df["Sequence 5\'"].apply(lambda x: tag + x)

df["Name with tag"] = df["Name"].apply(lambda x: "tag_" + x)
df.to_csv("final_primers_with_tag.csv", index=False)
df
```

```python
df = pd.read_csv("final_primers_withAllOld.csv")
df["Sequence 5\' with tag"] = df["Sequence 5\'"].apply(lambda x: tag + x)
df["Name with tag"] = df["Name"].apply(lambda x: "tag_" + x)
df.to_csv("final_primers_withAllOld_with_tag.csv", index=False)
df
```
