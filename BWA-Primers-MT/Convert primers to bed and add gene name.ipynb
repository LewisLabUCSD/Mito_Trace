```python
import pandas as pd 
from tqdm import tqdm
```

```python
def expand_anno_id(df, break_char="=",colname=8):
    df = df.copy()
    for ind, val in tqdm(df.iterrows()):
        curr = val[colname].split(';')

        for i in curr:

            v = i.strip().replace('"', "")
            if len(v) == 0:
                continue
            curr_split = v.split(break_char)

            df.at[ind, curr_split[0]] = curr_split[1]

    return df
```

```python
gff = pd.read_csv("MT_genome/MT.gene.gff",sep="\t", header=None)
gff = expand_anno_id(gff)
gff
```

## Convert to bed and add Gene Name

```python
f_name = "final_primers.csv"
primers_df = pd.read_csv(f_name)

primers_df["Gene"].apply(lambda x: gff[gff["ID"]==x]["Name"].values[0])
primers_df["Name"] = primers_df["Name"] + "_" + primers_df["Gene"].apply(lambda x: gff[gff["ID"]==x]["Name"].values[0])
primers_df["biotype"] = primers_df["Gene"].apply(lambda x: gff[gff["ID"]==x]["gene_biotype"].values[0]) + "_" + primers_df["Gene"].apply(lambda x: gff[gff["ID"]==x]["Name"].values[0])

primers_df["Chr"]  = "MT"
primers_df["Start"] = -1
primers_df["End"] = -1

for ind, val in primers_df.iterrows():
    if val["Strand"] == "-":
        primers_df.at[ind, "Start"]  = primers_df.loc[ind,"MT Position"] - len(primers_df.loc[ind,"Sequence 5'"])
        
        primers_df.at[ind, "End"] = primers_df.loc[ind,"MT Position"]
    else: 
        primers_df.at[ind, "Start"] = primers_df.loc[ind, "MT Position"] 
        primers_df.at[ind, "End"] = primers_df.loc[ind,"MT Position"] + len(primers_df.loc[ind,"Sequence 5'"])
primers_df["Score"] = 0
        
biotype = primers_df[primers_df["biotype"].str.contains("rRNA|tRNA")]
biotype[["Chr", "Start", "End", "biotype", "Score", "Strand"]].to_csv(f_name + "noncoding.bed", sep="\t", header=False, index=False)


primers_df[["Chr", "Start", "End", "Name", "Score", "Strand"]].to_csv(f_name + ".bed", sep="\t", header=False, index=False)

primers_df
```
