```python
from os.path import join
indirs = ["data/PBMC_P_cellSNP", "data/PBMC_J_cellSNP"]

variants = pd.read_csv(join(indirs[0], "cellSNP.base.vcf"), skiprows=1,sep="\t")
variants[indirs[0]] = variants.index.values+1
for ind, val in enumerate(indirs[1:]):
    curr = pd.read_csv(join(val, "cellSNP.base.vcf"), skiprows=1,sep="\t")
    curr[val] = curr.index.values+1
    variants = pd.merge(variants, curr,on=["#CHROM", "POS"], how="outer")

    
# Loop again and map the coordinates
vars_coords = dict()
full_vars = variants.copy()
full_vars["new ID"] = variants.index.values+1

# what the new index should be.
for val in indirs:
    curr_vars = pd.read_csv(join(val, "cellSNP.base.vcf"), skiprows=1, sep="\t")
    curr_vars = pd.merge(curr_vars, full_vars, how="inner")
    curr_vars.index = curr_vars.index+1
    vars_coords[val] = curr_vars["new ID"]  # Pandas series about

```
